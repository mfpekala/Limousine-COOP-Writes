```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
epsilon = 32
data = np.loadtxt(f'{epsilon}_A_paddingTops.txt')
data = data[data < 1000]
```

```python
# Plot a histogram of the data
plt.hist(data, bins=20)
plt.xticks(np.arange(0, 68, 12))
# Add vertical lines at the min and max
plt.axvline(data.min(), color='r', linestyle='dashed', linewidth=2, label=f'min = {data.min()}')
plt.axvline(data.max(), color='r', linestyle='dashed', linewidth=2, label=f'max = {data.max()}')
plt.axvline(epsilon, color='g', linestyle='solid', linewidth=2, label=f'epsilon = {epsilon}')
plt.legend()
plt.show()
```
