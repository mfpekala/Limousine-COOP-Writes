```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np
```

```python
# Results of the inserts_vs_index_power experiment

broad_df = pd.read_csv("shape/inserts_vs_index_power.csv")

xs = list(broad_df["data_size"].unique())
names = list(broad_df["name"].unique())
ymap = {}
for name in names:
    ymap[name] = np.array(broad_df[broad_df["name"] == name].groupby("data_size").mean(numeric_only=True).sort_values(by="data_size")["avg_node_size"])

for name in names:
    plt.plot(xs, ymap[name], label=name)
plt.xlabel("Total Data Size")
plt.ylabel("Data Indexed per Model (# entries)")
plt.title("Degradation of Indexing Power\n(initially 1M entries)")
plt.legend()
plt.show()
```

```python
# Results of the inserts_vs_index_power experiment

wlatency_neigh_df = pd.read_csv("time/inserts_vs_wlatency_neighbors.csv")

xs = list(wlatency_neigh_df["data_size"].unique())
names = list(wlatency_neigh_df["name"].unique())
ymap = {}
for name in names:
    ymap[name] = np.array(wlatency_neigh_df[wlatency_neigh_df["name"] == name].groupby("data_size").mean(numeric_only=True).sort_values(by="data_size")["time"])

for name in names:
    plt.plot(xs, ymap[name], label=name)
plt.xlabel("Total Data Size")
plt.ylabel("Time to insert 10K entries (ms)")
plt.title("Insert latency vs neighbors")
plt.legend()
plt.show()
```
